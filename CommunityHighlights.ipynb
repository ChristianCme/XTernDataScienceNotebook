# osXtern Community Highlights Analysis

Let's get started by import some necessary packages, open the data in pandas, and do some cleaning

```python
import sys
import csv

import sklearn
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
df = pd.read_csv('checkdataset.csv')
df.head()
```

```python
del df['Unnamed: 0']
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

## Visulaizations
Testing out graphs to see if there is any corrlations in the dataset

```python
x=df['xcoordinate']
y=df['ycoordinate']
plt.scatter(x,y)
```

Judging by purely the x and y coordnates, the users are choosing to share their location in very distinct locations

```python
z=df['timestamp']
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z,c=df['user'])
```

Even across two months, all the users are all grouping in same spots. And the colors show that while some users have prefered spots they go to other locations as well.

## Conclusions

So since the locations are pretty consistent across time, the users are primarily sharing data as a result of the location they are at or consistent events over the 2-month period, as opposed to sporadic events. Using a clustering algorithm the users could be categoized by the locations/events they share the most, and depending on what is at those locations, this could lead into some useful insights about our dempgraphic.
